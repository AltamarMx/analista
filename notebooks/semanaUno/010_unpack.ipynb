# Desempaca listas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def suma_tres(a,b,c):
    return a + b + c
```

```python
suma_tres(1,1,1)
```

```python
numeros = [1, 1, 1]
```

```python
suma_tres(numeros)
```

```python
suma_tres(*numeros)
```

```python
print(numeros)
```

```python
print(*numeros)
```

```python
f = "../../data/personas_cargas.csv"
horario = pd.read_csv(f,index_col=0,parse_dates=True)
horario
```

```python
people = [columna for columna in horario.columns if "People" in columna]
electricos = [columna for columna in horario.columns if "(W)" in columna]

people,electricos
```

```python
horario[electricos].dtypes
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.stackplot(horario.index,
             horario[electricos[0]],
             horario[electricos[1]],
             horario[electricos[2]],
             horario[electricos[3]],
             horario[electricos[4]],
             horario[electricos[5]],
             horario[electricos[6]],
             horario[electricos[7]],
             horario[electricos[8]],
             horario[electricos[9]],
#              data[electricos[8]],
             labels=electricos
            )


ax.legend(loc="upper left",
#           ncols=6,
          fontsize=6)
ax.grid()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.stackplot(horario.index,horario[electricos],labels=electricos)

ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.stackplot(horario.index,*[horario[electrico] for electrico in electricos],labels=electricos)

ax.legend()
plt.show()
```

```python
# En las nuevas versiones de matplotlib podemos no desempacar
fig, ax = plt.subplots(figsize=(12,6))

ax.stackplot(horario.index,[horario[electrico] for electrico in electricos],labels=electricos)

ax.legend()
```
