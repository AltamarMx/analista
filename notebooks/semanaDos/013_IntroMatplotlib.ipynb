```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
f = '../../data/Temixco_2018_10Min.csv'
tmx = pd.read_csv(f,index_col=0,parse_dates=True)
tmx.columns
```

```python
fig, ax = plt.subplots(figsize=(12,4),sharex=True)

# fecha1 = parse('2018-03-10')
# fecha2 = fecha1 + pd.Timedelta('3D')

ax.plot(tmx['Ig'],label='Ig')
ax.plot(tmx['Ib'],label='Ib')

ax.legend()
# ax.set_xlim(fecha1,fecha2)
```

```python
f = '../../data/personas_cargas.csv'
cargas = pd.read_csv(f,index_col=0,parse_dates=True)
columnas = cargas.columns
columnas
```

```python
etiquetas = [columna for columna in columnas if 'People' in columna]
etiquetas
```

```python
fig, ax = plt.subplots(figsize=(12,4))

ax.stackplot(cargas.index,[cargas[columna] for columna in columnas if 'People' in columna],labels= etiquetas)

ax.legend();
```
